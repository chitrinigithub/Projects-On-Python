```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

**LOADING THE FILE AND CONVERTING INTO DATAFRAME**

```python
df = pd.read_csv("/content/SampleSuperstore.csv")
df1 = pd.DataFrame(df)
print(df1)
```

**PULLING OUT THE COLUMNS FOR ANALYSIS**

```python
df1.columns
```

**CONVERTING THE NUMERICAL COLUMNS TO HISTOGRAM**

```python
df1.hist(bins=50, figsize=(11,8))
```

**ANALYSIS by USING REGRESSION PLOT TO CHECK SALES AND PROFIT ARE CORRELATED**

```python
sns.regplot(x="Sales", y="Profit", data = df1) #Sales and Profit are positively correlated 
```

**ANALYSIS by USING REGRESSION PLOT TO CHECK SALES AND QUANTITY ARE CORRELATED**

****

```python
sns.regplot(x="Sales", y="Quantity", data = df1) # Sales and Quantity are positively correlated
```

**ANALYSIS by USING REGRESSION PLOT TO CHECK PROFIT AND QUANTITY ARE CORRELATED**

```python
sns.regplot(x="Profit", y="Quantity", data = df1) #Profit and Quantity are positively correlated
```

**ANALYSIS by USING REGRESSION PLOT TO CHECK SALES AND DISCOUNT ARE CORRELATED**

```python
sns.regplot(x="Sales", y="Discount", data = df1)#Sales and Discount are negatively correlated 
```

**ANALYSIS by USING REGRESSION PLOT TO CHECK DISCOUNT AND PROFIT ARE CORRELATED**

```python
sns.regplot(x="Profit", y="Discount", data = df1)#Profit and Discount are negatively correlated
```

**SEGMENTWISE PROFIT IN EACH CATEGORY**

```python
sns.barplot(data=df1, x='Segment', y='Profit', hue='Category')
#In all three segments Technology is making the highest profit
```

**REGIONWISE SALES**

```python
sns.barplot(data=df1, x='Sales', y='Region', hue='Country')
#South region is having highest almost fairly similar sales to that of east region
#Second highest region of sales being West
#Third Highest being Central in Sales
```

**REGIONWISE PROFIT**

```python
sns.barplot(data=df1, x='Profit', y='Region', hue='Country')
#Highest profit is in West and East region
#Second Highest profit is in South region
#Third highest is in Central region
```

**ANALYSIS USING BOX PLOT TO CHECK WHETHER SHIP MODE AFFECTS THE SALES**

```python
df1.boxplot(column="Sales",by="Ship Mode",figsize=(20,9))
#The box plot shows all the boxes lies in the same line without any change in position which shows there is no effect of shipmode on sales
```

**ANALYSIS USING BOX PLOT TO CHECK WHETHER SHIP MODE AFFECTS THE PROFIT**

```python
df1.boxplot(column="Profit",by="Ship Mode",figsize=(20,9))
#The box plot shows all the boxes lies in the same line without any change in position which shows there is no effect of shipmode on profit
```
