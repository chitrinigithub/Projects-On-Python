```python
#Problem-6-User defined function to calculate median of the dataset wherever required
def my_median(sample):
  n = len(sample)                                   #Finding out the length of the sample and storing it in n variabe
  index = n // 2                                    #Getting  the middle element of the sample by diving the length of sample by 2 and storing it in index variable
    # Sample with an odd number of observations
  if n % 2:                              #If n mod 2 has a remainder I am returning the middle element of the sample by arranging the sample data in ascending order using sorted()
   return sorted(sample)[index]
  # Sample with an even number of observations
  return sum(sorted(sample)[index - 1:index + 1]) / 2      #else I am returning sorted sample where i am finding the two middle elements in the list and returning their average


new_df=my_median(df_1["median_income"])  #Here i am passing the values for column median_income of df_1 and storing it in new_df 
new_df
```
